```python
def save_file(filepath, content):
  with open(filepath, 'w', encoding='utf-8') as output_file:
    output_file.write(content)

def convert_pdf2txt_by_pdfplumber(src_dir, dest_dir):
  files = os.listdir(src_dir)
  files = [f for f in files if '.pdf' in f]
  print(f'files: {files}')
  for filename in files:
    try:
      fullpath = os.path.join(src_dir,filename)
      with pdfplumber.open(fullpath) as pdf:
        output = ''
        for page_number, page in enumerate(pdf.pages):
          output += page.extract_text()
          output += f'\n\nPAGE {page_number}\n'
        save_file(os.path.join(dest_dir,filename.replace('.pdf','.txt')), output.strip())
    except Exception as oops:
      print(oops, ',', filename)

def convert_pdf2txt_by_pymupdf(src_dir, dest_dir):
  files = os.listdir(src_dir)
  files = [f for f in files if '.pdf' in f]
  print(f'files: {files}')
  for filename in files:
    try:
      fullpath = os.path.join(src_dir,filename)
      with fitz.open(fullpath) as doc:
        output = ''
        for page_number, page in enumerate(doc):
          output += page.get_text()
          output += f'\n\nPAGE {page_number}\n'
      save_file(os.path.join(dest_dir,filename.replace('.pdf','.txt')), output.strip())
    except Exception as oops:
      print(oops, ',', filename)
```

```python
import fitz

# open the PDF file using PyMuPDF
pdf_file = "example.pdf"
doc = fitz.open(pdf_file)

# loop through the pages of the PDF
for page in doc:
    # get the page text blocks
    blocks = page.get_text("dict")["blocks"]
    
    # determine the column boundaries
    column1_left = page.rect.width / 2
    column2_right = page.rect.width / 2
    
    # loop through the text blocks and assign them to columns
    column1 = []
    column2 = []
    for block in blocks:
        bbox = block["bbox"]
        text = block["text"]
        if bbox[2] < column1_left:
            column1.append(text)
        elif bbox[0] > column2_right:
            column2.append(text)
        else:
            # handle blocks that overlap both columns
            pass
    
    # print the two columns for this page
    print("Column 1:")
    print("\n".join(column1))
    print("\n\nColumn 2:")
    print("\n".join(column2))
```
